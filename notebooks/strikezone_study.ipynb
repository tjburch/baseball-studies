```python
#fig, axobj = plt.subfigure(ncols=4,nrows=3)
from collections import OrderedDict
result_dictionary = OrderedDict()
for strikes in range(0,3):
    for balls in range(0,4):
        
        # Get count status
        this_df = data.query("strikes == %i and balls == %i" % (strikes, balls))
        
        # Get strikes only
        this_df = this_df.query("type == 'S'")
        
        strike_upper_x =  this_df.plate_x.quantile(q=.95)
        strike_lower_x =  this_df.plate_x.quantile(q=.05)
        strike_upper_z =  this_df.plate_z.quantile(q=.95)
        strike_lower_z =  this_df.plate_z.quantile(q=.05)
        
        result_dictionary[(balls,strikes)] = [strike_upper_x, strike_lower_x, strike_upper_z, strike_lower_z]
```

```python
fig = plt.figure()
ax = plt.gca()
idx = 0
colors = plt.cm.jet(np.linspace(0,1,len(result_dictionary)))
color_dict = {} # For consistent coloring
for key in result_dictionary:
    # unpack
    xmax = result_dictionary[key][0]
    xmin = result_dictionary[key][1]
    zmax= result_dictionary[key][2]
    zmin= result_dictionary[key][3]

    plt.hlines([zmin,zmax], xmax=xmax, xmin=xmin, color=colors[idx])
    plt.vlines([xmin,xmax], ymax=zmax, ymin=zmin, color=colors[idx], label=key)
    color_dict[key] = colors[idx]
    idx += 1
    
plt.xlim((-3,3))
plt.ylim((0,5))
plt.legend(title="Count")
```

```python
extremum = [(0,0), (0,2),(3,0),(3,2)]
for key in extremum:
    # unpack
    xmax = result_dictionary[key][0]
    xmin = result_dictionary[key][1]
    zmax= result_dictionary[key][2]
    zmin= result_dictionary[key][3]

    plt.hlines([zmin,zmax], xmax=xmax, xmin=xmin, color=color_dict[key])
    plt.vlines([xmin,xmax], ymax=zmax, ymin=zmin, color=color_dict[key], label=key)
    idx += 1
    
plt.xlim((-3,3))
plt.ylim((0,5))
plt.legend(title="Count")
```
